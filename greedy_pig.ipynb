```julia
function generate_data(num_rounds=10000)
    vec = Float64[]
    for i in 1:50
        #roll dice note down value v1
        #if v1 is 1, you get 0. go to next round
        #if v1 is not 1 but less than i, roll again and note down value v2.
        #if v2 is 1, you get 0. go to next round.
        #else if v2+v1 is less than i, roll again and note down value v3.If v2+v1 is >=i 
        #then you get v2+v1 for the current round and you go to next round. 
        #you add up values of all rounds and take their average. 
        #Do this for all i's.
        round = 1
        sum=0
        while round<=num_rounds
            round_value=0
            while round_value<i
                rnum=rand(1:6)
                if rnum==1
                    round_value=0
                    break
                else
                    round_value+=rnum
                end
            end
            sum+=round_value
            round+=1
        end
        append!(vec, sum/num_rounds)
    end
    return vec
end
```

```julia
data=generate_data(1000000)
```
